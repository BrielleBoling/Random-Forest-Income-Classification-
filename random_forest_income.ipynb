```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree, export_text
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#read file and get basic info

myIncome = pd.read_csv('Income2.csv')
myIncome.info()
myIncome
```

```python
#RESAMPLING - creating more samples

newSamples = myIncome.sample(n=10000, replace=True)
newSamples
```

DATA EXPLORATION

```python
# Question 5a and 5b
myIncome.info()

# Quesiton 5c
uniqueCreditScores = myIncome['CreditScore'].nunique()
print(uniqueCreditScores)

# Quesiton 5d - number of people with an MBA
mba = (myIncome['Education'] == 'MBA').sum()
print("There are ", mba, " people with an MBA")

# Question 5e - how many features have missing values
missingValues = myIncome.isnull().sum()
print("There are ",missingValues, " features with missing values")
```

```python
#creating dependent and independent variables

y = np.array(newSamples['Education'])
x = newSamples[['Experience','GPA','Age','Income','CreditScore','NoCars','ParkingTickets','NewestCarYear','NumberKids']]

#split data into testing and training data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#Building forest

bagging_MyTree = BaggingClassifier(DecisionTreeClassifier(criterion= 'gini'), n_estimators = 100, max_samples = 1.0, max_features= 3, random_state=42)
bagging_MyTree.fit(x_train, y_train)

#Train the model
bagging_MyTree.fit(x_train, y_train)

#Model predictions
y_pred_bag = bagging_MyTree.predict(x_test)
```

```python
# Evaluating the first forest
accuracy_bag = metrics.accuracy_score(y_test, y_pred_bag)
print("Accuracy:", accuracy_bag)

conf_matrix = metrics.confusion_matrix(y_test, y_pred_bag)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
#all five measures except accuracy as already calculated
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

precision_bag = precision_score(y_test, y_pred_bag, average='weighted')
recall_bag = recall_score(y_test, y_pred_bag, average='weighted')
f1_bag = f1_score(y_test, y_pred_bag, average='weighted')

print("Precision:", precision_bag)
print("Recall:", recall_bag)
print("F1 Score:", f1_bag)
```

```python
# second random forest

myRF = RandomForestClassifier(n_estimators=500, random_state=42)
myRF.fit(x_train, y_train)

#Model predictions
y_pred_rf = myRF.predict(x_test)

#accuracy of random forest
accuracy_rf = metrics.accuracy_score(y_test, y_pred_rf)
print("Accuracy:", accuracy_rf)

conf_matrix = metrics.confusion_matrix(y_test, y_pred_rf)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
#all five measures

precision_rf = precision_score(y_test, y_pred_rf, average='weighted')
recall_rf = recall_score(y_test, y_pred_rf, average='weighted')
f1_rf = f1_score(y_test, y_pred_rf, average='weighted')

print("Precision:", precision_rf)
print("Recall:", recall_rf)
print("F1 Score:", f1_rf)
```
